#Criar uma Calculadora Simples

```python
#entrada
a = int(input("digite o primeiro numero: "))
b = int(input("digite o segundo numero: "))
#processamento
soma = a + b
subtracao = a - b
divisao = a / b
multiplicacao = a * b
#saida
print ("------------------- CALCULADORA SIMPLES -----------------")
print ("")
print ("-------------------------- Soma -------------------------")
print (f"a soma de {a} + {b} é igual a: {soma:.2f}")
print ("")
print ("------------------------ Subtração ----------------------")
print (f"a subtração de {a} - {b} é igual a: {subtracao:.2f}")
print ("")
print ("---------------------- Multiplicação --------------------")
print (f"a multiplicação de {a} * {b} é igual a: {multiplicacao:.2f}")
print ("")
print ("------------------------- Divisão -----------------------")
print (f"a divisão de {a} / {b} é igual a: {divisao:.2f}")
```

# pode se utilizar no lugar da divisão:

* a // b
  * Para se calcular o valor inteiro mais aproximado

* a % b
  * Para se calcular o valor do resto da divisão
